```python
# Importing the libraries and packages required for analyzing data and ploting graphs.
import pandas as pd
%matplotlib inline
from IPython.display import HTML,SVG
import pygal
import seaborn as sns
import numpy as np
```

```python
# Importing the dataset using pandas read Api.
iris = pd.read_csv("Iris.csv")
```

```python
#Shows the data in the data sets.
iris.head()
```

```python
# In column species, it shows how many types of species are present and its data count.
iris["Species"].value_counts()
```

```python
# Analysing the cout,mean,max,min etc. for all columns available.
iris.describe()
```

```python
# Extracting the data for each species seperatly
# Mapping all the data of each species seperatly, 
#by checking the string name and storing it in, from column species.

Setosa = iris['Species'].map(lambda x: x.startswith('Iris-setosa'))
Versicolor = iris['Species'].map(lambda x: x.startswith('Iris-versicolor'))
Virginica = iris['Species'].map(lambda x: x.startswith('Iris-virginica'))
```

```python
# Making tables for each species.
setosa = iris[Setosa]
versicolor = iris[Versicolor]
virginica = iris[Virginica]
```

```python
# Adding html script for html_pygal api, this is only required in python 3.0
html_pygal = """
<!DOCTYPE html>
<html>
  <head>
  <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
  <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/pygal-tooltips.js"></script>
    <!-- ... -->
  </head>
  <body>
    <figure>
      {pygal_render}
    </figure>
  </body>
</html>
"""
```

```python
# Ploting a Bar graph showing number of data available for each species in the data set.

bar = pygal.Bar(print_labels=True,print_values=True)
bar.title = 'Analyzing Data of Species'
bar.x_labels = map (str,('Setosa', 'Versicolor','Virginica'))
bar.add('Iris Data Set', iris["Species"].value_counts())
bar.render_to_file('Species_data_set_count.svg')
HTML(html_pygal.format(pygal_render=bar.render()))
```

```python
# Analysing Specie Setosa data sets and ploting a piechart using pygal.
pie_chart = pygal.Pie(print_labels=True,print_values=True)
pie_chart.title = 'Setosa Attributes'
pie_chart.add('SepalLengthCm', setosa['SepalLengthCm'])
pie_chart.add('SepalWidthCm', setosa['SepalWidthCm'])
pie_chart.add('PetalLengthCm', setosa['PetalLengthCm'])
pie_chart.add('PetalWidthCm', setosa['PetalWidthCm'])
pie_chart.render_to_file('Setosa_Attributes.svg')
HTML(html_pygal.format(pygal_render=pie_chart.render()))
```

```python
# Analysing Specie Versicolor data sets and ploting a piechart using pygal.
pie_chart = pygal.Pie(print_labels=True,print_values=True)
pie_chart.title = 'Versicolor Attributes'
pie_chart.add('SepalLengthCm', versicolor['SepalLengthCm'])
pie_chart.add('SepalWidthCm', versicolor['SepalWidthCm'])
pie_chart.add('PetalLengthCm', versicolor['PetalLengthCm'])
pie_chart.add('PetalWidthCm', versicolor['PetalWidthCm'])
pie_chart.render_to_file('Versicolor_Attributes.svg')
HTML(html_pygal.format(pygal_render=pie_chart.render()))
```

```python
# Analysing Specie Virginica data sets and ploting a piechart using pygal.
pie_chart = pygal.Pie(print_labels=True,print_values=True)
pie_chart.title = 'Virginica Attributes'
pie_chart.add('SepalLengthCm', virginica['SepalLengthCm'])
pie_chart.add('SepalWidthCm', virginica['SepalWidthCm'])
pie_chart.add('PetalLengthCm', virginica['PetalLengthCm'])
pie_chart.add('PetalWidthCm', virginica['PetalWidthCm'])
pie_chart.render_to_file('Virginica_Attributes.svg')
HTML(html_pygal.format(pygal_render=pie_chart.render()))
```

```python
# Analysing SepalLengthCm of each Species seperately

bar = pygal.StackedBar(print_labels=True)
bar.title = "Sepal Length Analysis of each Species. in (cm)"
bar.add ('Setosa',setosa['SepalLengthCm'])
bar.add ('Virginica',virginica['SepalLengthCm'])
bar.add ('Versicolor',versicolor['SepalLengthCm'])
bar.render_to_file('Sepal_Length_Analysis.svg')
HTML(html_pygal.format(pygal_render=bar.render()))
```

```python
# Taking the average of all the data of sepal length of all species.

Average_Setosa_length = setosa["SepalLengthCm"].mean()
Average_Virginica_length = virginica["SepalLengthCm"].mean()
Average_Versicolor_length = versicolor["SepalLengthCm"].mean()
```

```python
# Ploting the min average of all the Iris Species sepal length 
bar = pygal.Bar(print_labels=True,print_values=True)
bar.title = "Sepal Length Average of each Species. in (cm)"
bar.add ('Setosa',Average_Setosa_length)
bar.add ('Virginica',Average_Virginica_length)
bar.add ('Versicolor',Average_Versicolor_length)
bar.render_to_file('Sepal_Length_Average.svg')
HTML(html_pygal.format(pygal_render=bar.render()))
```

```python
# Analysing SepalWidthCm of each Species seperately

line_chart = pygal.Line(print_labels=True)
line_chart.title = "Sepal Width Analysis of each Species. in (cm)"
line_chart.x_labels = map(str, range(0,50))
line_chart.add ('Setosa',setosa['SepalWidthCm'])
line_chart.add ('Virginica',virginica['SepalWidthCm'])
line_chart.add ('Versicolor',versicolor['SepalWidthCm'])
line_chart.render_to_file('Sepal_Width_Analysis.svg')
HTML(html_pygal.format(pygal_render=line_chart.render()))
```

```python
# Taking the average of all the data of sepal Width of all species.

Average_Setosa_width = setosa["SepalWidthCm"].mean()
Average_Virginica_width = virginica["SepalWidthCm"].mean()
Average_Versicolor_width = versicolor["SepalWidthCm"].mean()
```

```python
# Ploting the min average of all the Iris Species sepal Width 
bar = pygal.HorizontalBar(print_labels=True,print_values=True)
bar.title = "Sepal Width Average of each Species. in (cm)"
bar.add ('Setosa',Average_Setosa_width)
bar.add ('Virginica',Average_Virginica_width)
bar.add ('Versicolor',Average_Versicolor_width)
bar.render_to_file('Sepal_Width_Average.svg')
HTML(html_pygal.format(pygal_render=bar.render()))
```

```python
# Analysing PetalLengthCm of each Species seperately

line_chart = pygal.StackedLine(print_labels=True)
line_chart.title = "Petal Length Analysis of each Species. in (cm)"
line_chart.x_labels = map(str, range(0,50))
line_chart.add ('Setosa',setosa['PetalLengthCm'])
line_chart.add ('Virginica',virginica['PetalLengthCm'])
line_chart.add ('Versicolor',versicolor['PetalLengthCm'])
line_chart.render_to_file('Petal_Length_Analysis.svg')
HTML(html_pygal.format(pygal_render=line_chart.render()))
```

```python
# Taking the average of all the data of Petal length of all species.
Average_Setosa_petal_len = setosa["PetalLengthCm"].mean()
Average_Virginica_petal_len = virginica["PetalLengthCm"].mean()
Average_Versicolor_petal_len = versicolor["PetalLengthCm"].mean()
```

```python
# Ploting the min average of all the Iris Species Petal length 
bar = pygal.StackedBar(print_labels=True,print_values=True)
bar.title = "Pepal Length Average of each Species. in (cm)"
bar.add ('Setosa',Average_Setosa_petal_len)
bar.add ('Virginica',Average_Virginica_petal_len)
bar.add ('Versicolor',Average_Versicolor_petal_len)
bar.render_to_file('Petal_Length_Average.svg')
HTML(html_pygal.format(pygal_render=bar.render()))
```

```python
# Analysing PetalWidthCm of each Species seperately

radar_chart  = pygal.Radar(print_labels=True)
radar_chart .title = "Petal Width Analysis of each Species. in (cm)"
radar_chart .x_labels = map(str, range(0,50))
radar_chart .add ('Setosa',setosa['PetalWidthCm'])
radar_chart .add ('Virginica',virginica['PetalWidthCm'])
radar_chart .add ('Versicolor',versicolor['PetalWidthCm'])
radar_chart.render_to_file('Petal_Width_Analysis.svg')
HTML(html_pygal.format(pygal_render=radar_chart .render()))
```

```python
# Taking the average of all the data of Petal Width of all species.
Average_Setosa_petal_Width = setosa["PetalWidthCm"].mean()
Average_Virginica_petal_Width = virginica["PetalWidthCm"].mean()
Average_Versicolor_petal_Width = versicolor["PetalWidthCm"].mean()
```

```python
# Ploting the min average of all the Iris Species Petal Width 
dot_chart = pygal.Dot(print_labels=True,print_values=True)
dot_chart.title = "Petal Width Average of each Species. in (cm)"
dot_chart.add ('Setosa',Average_Setosa_petal_Width)
dot_chart.add ('Virginica',Average_Virginica_petal_Width)
dot_chart.add ('Versicolor',Average_Versicolor_petal_Width)
dot_chart.render_to_file('Petal_Width_Average.svg')
HTML(html_pygal.format(pygal_render=dot_chart.render()))
```

```python
# This Analysis will help us to understand, how to identify a particular species.
#Below is the Analysis of Setosa.
gauge = pygal.SolidGauge(inner_radius=0.50,outer_radius = 0.10)
gauge.title = "This Analysis tells, How to identify, if its Setosa."
gauge.legend_at_bottom = True
gauge.add('SepalLengthCm min',setosa["SepalLengthCm"].min())
gauge.add('SepalLengthCm max',setosa["SepalLengthCm"].max())
gauge.add('SepalWidthCm min',setosa["SepalWidthCm"].min())
gauge.add('SepalWidthCm max',setosa["SepalWidthCm"].max())
gauge.add('PetalLengthCm min',setosa["PetalLengthCm"].min())
gauge.add('PetalLengthCm max',setosa["PetalLengthCm"].max())
gauge.add('PetalWidthCm min',setosa["PetalWidthCm"].min())
gauge.add('PetalWidthCm max',setosa["PetalWidthCm"].max())
gauge.render_to_file('Setosa_Overall_Analysis.svg')
HTML(html_pygal.format(pygal_render=gauge.render()))
```
